```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt  
import pandas_datareader.data as web
#Group 21: Broad Effects of Covid-19 on Tourism and Transportation in the U.S.
#Lorenzo Puglielli, Lucy Kuo, Trevor Wasbotten, Sophia Bergman
```

```python
year = pd.read_excel('year.xlsx')
year['time'] = pd.to_datetime(year['time']) #set time column to datetime
year.set_index('time', inplace=True) #Set time column to index
year
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(year.index, year['Northeast'], color='red', label = 'Northeast', alpha=1)
ax.plot(year.index, year['Midwest'], color='blue', label = 'Midwest', alpha=1)
ax.plot(year.index, year['South'], color='green', label = 'South', alpha=1)
ax.plot(year.index, year['West'], color='purple', label = 'West', alpha=1)
ax.plot(year.index, year['United States'], color='black', label = 'United States', alpha=1)
#Plot each of the columns against the index

ax.set_ylabel('Percent Change')
ax.set_title('YEAR-OVER-YEAR % CHANGE IN WEEKLY TRAVEL SPENDING')

ax.set_xticks(['2020-11-14', '2020-11-21', '2020-11-28', '2020-12-05', '2020-12-12', '2020-12-19', '2020-12-26', '2021-01-02'])
#Set list of dates as marks on x-axis

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend(frameon=False)
#Set legend

plt.savefig('year_project.svg')
plt.show()
```
