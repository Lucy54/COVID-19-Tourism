```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lv = pd.read_excel('Year_to_Date_Summary_for_2020_Revised.xlsx'
                  , header=6)

for col in lv.columns:
    if 'Unnamed' in str(col):
        lv.drop(columns=col, inplace=True)

lv
```

```python
vol = lv.iloc[[0]]
vol = vol.transpose()
vol.columns = vol.iloc[0]
vol = vol[1:]
vol['growth'] = vol['Visitor Volume'].pct_change() * 100
vol
```

```python
use = vol[:-1]
```

```python
fix, ax = plt.subplots(figsize=(10, 5))

ax.plot(use.index, use['Visitor Volume'])

ax.set_ylabel('Number of Visitors')
ax.set_title('Las Vegas Visitors')

sns.despine()

plt.savefig('lvvisitors.svg')
plt.show()
```

```python
fix, ax = plt.subplots(figsize=(10, 5))

ax.plot(use.index, use['growth'])

ax.set_ylabel('Percent Change')
ax.set_title('Las Vegas Visitors Growth Rate')

sns.despine()

plt.savefig('lvgrowth.svg')
plt.show()
```
